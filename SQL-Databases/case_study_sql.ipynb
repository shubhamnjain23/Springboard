Q10: Produce a list of facilities with a total revenue less than 1000.

```python
query10 = """
SELECT name, SUM(
CASE WHEN b.memid =0
THEN slots * guestcost
ELSE slots * membercost
END ) AS revenue
FROM Bookings b
INNER JOIN Facilities f ON b.facid = f.facid
GROUP BY f.facid
HAVING revenue <1000
ORDER BY revenue;
"""
```

```python
df10 = pd.read_sql_query(query10, engine)
df10
```

Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
query11 = """
SELECT m.firstname || ' ' || m.surname AS member, m2.firstname || ' ' || m2.surname AS recommendedby
FROM Members m
JOIN Members m2 ON m2.memid = m.recommendedby
WHERE m.recommendedby >0
ORDER BY m.firstname, m.surname;
"""
```

```python
df11 = pd.read_sql_query(query11, engine)
df11
```

Q12: Find the facilities with their usage by member, but not guests

```python
query12 = """
SELECT f.name, COUNT(b.bookid) as member_usage
FROM Bookings b
JOIN Facilities f 
ON f.facid = b.facid 
WHERE b.memid > 0 
GROUP by f.facid ;
"""
```

```python
df12 = pd.read_sql_query(query12, engine)
df12
```

Q13: Find the facilities usage by month, but not guests

```python
query13 = """
SELECT strftime('%m', starttime) as month,count(b.facid) as facid_count, name
from Bookings b
inner join Facilities f
on b.facid = f.facid
GROUP by Month, name;
"""
```

```python
df13 = pd.read_sql_query(query13, engine)
df13
```
